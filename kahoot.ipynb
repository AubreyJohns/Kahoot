# Old Kahoot


[Kahoot Wikipedia](https://en.wikipedia.org/wiki/Kahoot!)

[Kahoot](https://kahoot.com/) is a game-based learning platform, used as educational technology in schools and other educational institutions. Its learning games, "Kahoots", are user-generated multiple-choice quizzes that can be accessed via a web browser or the Kahoot app.

Kahoot! can be used to review students' knowledge, for formative assessment, or as a break from traditional classroom activities.Kahoot! also includes trivia quizzes.

Kahoot simulation to determine the most important quality.
- Response time: The rate of answering questions
- Streak: The number of consecutive correct answers
- Accuracy: The sum of correct answers

#### Questions
Whether faster response correlates to higher total points
Whether higher streaks thus higher bonuses leads to higher total points

#### Assumptions
All questions have same time of 30s


```python
old_header_list = ["Points", "Bonuses", "Accuracy", "Total Response Time"]
old_df = pd.read_csv('old_kahoot_data.csv', header=None, skiprows=1, names=old_header_list, low_memory=False)
old_df.head()
```

```python
# Scatter plots
```

```python
sns.lmplot(x='Bonuses', y='Points', data=old_df)
```

```python
sns.lmplot(x='Accuracy', y='Points', data=old_df)
```

```python
sns.regplot(x='Total Response Time', y='Points', data=old_df)
```

```python
# Calculate correlations
corr = old_df.corr()

# Heatmap
sns.heatmap(corr, annot=True, cmap="coolwarm")
```

```python
# because response time varies on different questions the correlation is inaccurate
```
